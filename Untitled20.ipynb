```python
# Full preprocessing script (Colab / local)
# -----------------------------------------------------------------------------
# If running in Colab, upload all CSVs to Files first. Then run this cell.
# -----------------------------------------------------------------------------

# Install imblearn if missing (Colab)
try:
    import imblearn
except Exception:
    !pip install -q imbalanced-learn

# ======================
# IMPORTS
# ======================
import os
import joblib
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

from imblearn.over_sampling import SMOTE

# ======================
# 1) LOAD RAW CSVs (update paths if needed)
# ======================
train_df = pd.read_csv("train.csv")                         # labeled
campaign_df = pd.read_csv("campaign_data.csv")
coupon_item_mapping = pd.read_csv("coupon_item_mapping.csv")
item_data = pd.read_csv("item_data.csv")
cust_demo = pd.read_csv("customer_demographics.csv")
cust_trans = pd.read_csv("customer_transaction_data.csv")
test_df = pd.read_csv("test.csv")                           # unlabeled

print("Files loaded.")

# ======================
# 2) CLEAN & FEATURE ENGINEER (campaign, coupon, customer, transactions)
# ======================
# Campaign: parse and duration
campaign_df = campaign_df.copy()
campaign_df['start_date'] = pd.to_datetime(campaign_df['start_date'], format="%d/%m/%y", errors='coerce')
campaign_df['end_date']   = pd.to_datetime(campaign_df['end_date'],   format="%d/%m/%y", errors='coerce')
campaign_df['campaign_duration'] = (campaign_df['end_date'] - campaign_df['start_date']).dt.days

# Coupon-item aggregates (coupon-level features)
coupon_item_merged = coupon_item_mapping.merge(item_data, on='item_id', how='left')
coupon_item_features = coupon_item_merged.groupby('coupon_id').agg(
    num_unique_items=('item_id','nunique'),
    num_unique_brands=('brand','nunique'),
    num_unique_categories=('category','nunique'),
    prop_local=('brand_type', lambda x: (x == 'Local').mean()),
    prop_established=('brand_type', lambda x: (x == 'Established').mean()),
    top_category=('category', lambda x: x.mode()[0] if not x.mode().empty else 'Unknown')
).reset_index()

# Customer demographics: coerce numeric fields & fill missing
cust_demo = cust_demo.copy()
if 'family_size' in cust_demo.columns:
    cust_demo['family_size'] = pd.to_numeric(cust_demo['family_size'], errors='coerce')
if 'no_of_children' in cust_demo.columns:
    cust_demo['no_of_children'] = pd.to_numeric(cust_demo['no_of_children'], errors='coerce').fillna(0)
if 'marital_status' in cust_demo.columns:
    cust_demo['marital_status'] = cust_demo['marital_status'].fillna('Unknown')

# Customer transactions -> aggregated features
cust_trans = cust_trans.copy()
cust_trans['date'] = pd.to_datetime(cust_trans['date'], errors='coerce')
# discounts stored negative in raw; make absolute
if 'other_discount' in cust_trans.columns:
    cust_trans['other_discount'] = cust_trans['other_discount'].abs()
if 'coupon_discount' in cust_trans.columns:
    cust_trans['coupon_discount'] = cust_trans['coupon_discount'].abs()

cust_trans_features = cust_trans.groupby('customer_id').agg(
    total_quantity=('quantity','sum'),
    total_spend=('selling_price','sum'),
    avg_spend=('selling_price','mean'),
    total_other_discount=('other_discount','sum'),
    total_coupon_discount=('coupon_discount','sum'),
    unique_items=('item_id','nunique'),
    last_purchase=('date','max'),
    num_transactions=('date','count')
).reset_index()

# recency
max_date = cust_trans['date'].max()
cust_trans_features['recency_days'] = (max_date - cust_trans_features['last_purchase']).dt.days
cust_trans_features = cust_trans_features.drop(columns=['last_purchase'])

# ======================
# 3) MERGE to form train and test feature tables
# ======================
# Merge labeled train
train_merged = (train_df
    .merge(campaign_df, on='campaign_id', how='left')
    .merge(coupon_item_features, on='coupon_id', how='left')
    .merge(cust_demo, on='customer_id', how='left')
    .merge(cust_trans_features, on='customer_id', how='left')
)

# Drop original row id if present (we'll keep test ids separately)
if 'id' in train_merged.columns:
    train_merged = train_merged.drop(columns=['id'])

# Merge test (preserve id)
test_merged = (test_df
    .merge(campaign_df, on='campaign_id', how='left')
    .merge(coupon_item_features, on='coupon_id', how='left')
    .merge(cust_demo, on='customer_id', how='left')
    .merge(cust_trans_features, on='customer_id', how='left')
)
test_ids = test_merged['id'].values if 'id' in test_merged.columns else None

print("Merged shapes -> train:", train_merged.shape, "test:", test_merged.shape)

# ======================
# 4) SELECT features (drop keys and raw dates)
# ======================
drop_cols = ['campaign_id','coupon_id','customer_id','start_date','end_date']
X_all = train_merged.drop(columns=[c for c in drop_cols if c in train_merged.columns] + ['redemption_status'])
y_all = train_merged['redemption_status'].astype(int).values

X_test_all = test_merged.copy()
if 'id' in X_test_all.columns:
    X_test_all = X_test_all.drop(columns=[c for c in drop_cols if c in X_test_all.columns])
else:
    X_test_all = X_test_all.drop(columns=[c for c in drop_cols if c in X_test_all.columns])

# ======================
# 5) Build preprocessing pipeline (impute → scale → one-hot)
# ======================
# Identify numeric & categorical columns (from X_all DataFrame)
num_cols = X_all.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X_all.select_dtypes(include=['object','category']).columns.tolist()

print("Numeric columns:", len(num_cols))
print("Categorical columns:", len(cat_cols))

# numeric pipeline: impute (median) + scale
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# categorical pipeline: impute (most_frequent) + onehot
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
], remainder='drop')

# Fit preprocessor on labeled data (this will impute missing values)
X_proc = preprocessor.fit_transform(X_all)
X_test_proc = preprocessor.transform(X_test_all)

# Build readable feature names
feature_names = []
feature_names.extend(num_cols)
if len(cat_cols) > 0:
    ohe = preprocessor.named_transformers_['cat'].named_steps['onehot']
    ohe_names = list(ohe.get_feature_names_out(cat_cols))
    feature_names.extend(ohe_names)

print("Processed features count:", len(feature_names))

# Save preprocessor
joblib.dump(preprocessor, "preprocessor.joblib")
print("Saved preprocessor -> preprocessor.joblib")

# Verify no NaNs remain after imputation
print("Any NaN in train processed?", np.isnan(X_proc).any())
print("Any NaN in test processed?", np.isnan(X_test_proc).any())

# ======================
# 6) TRAIN/VAL SPLIT (stratified)
# ======================
X_train_proc, X_val_proc, y_train, y_val = train_test_split(
    X_proc, y_all, test_size=0.2, stratify=y_all, random_state=42
)

print("Shapes before balancing: train:", X_train_proc.shape, "val:", X_val_proc.shape)
print("Label dist train:", np.bincount(y_train), "val:", np.bincount(y_val))

# ======================
# 7) BALANCE using SMOTE -> to 1:1 ratio (equal 0 and 1)
#    automatically adjust k_neighbors if minority small
# ======================
def smote_balance(X, y, sampling_strategy=1.0, random_state=42):
    # X: numpy array (dense)
    # y: 1D array
    unique, counts = np.unique(y, return_counts=True)
    if len(unique) == 1:
        return X, y  # nothing to do
    minority_count = counts.min()
    # choose k_neighbors <= minority_count - 1, but at least 1
    k_neighbors = min(5, max(1, minority_count - 1))
    sm = SMOTE(sampling_strategy=sampling_strategy, random_state=random_state, k_neighbors=k_neighbors)
    X_res, y_res = sm.fit_resample(X, y)
    return X_res, y_res

# Balance training (majority/ minority -> 1:1 sampling_strategy=1.0)
X_train_bal, y_train_bal = smote_balance(X_train_proc, y_train, sampling_strategy=1.0, random_state=42)
print("After SMOTE train class counts:", np.bincount(y_train_bal))

# Optionally balance validation too (set to True if you want val balanced)
BALANCE_VAL = True
if BALANCE_VAL:
    X_val_bal, y_val_bal = smote_balance(X_val_proc, y_val, sampling_strategy=1.0, random_state=24)
    print("After SMOTE val class counts:", np.bincount(y_val_bal))
else:
    X_val_bal, y_val_bal = X_val_proc, y_val

# ======================
# 8) SAVE balanced CSVs and test features
# ======================
# Train balanced DataFrame (with readable column names)
df_train_bal = pd.DataFrame(X_train_bal, columns=feature_names)
df_train_bal['redemption_status'] = y_train_bal.astype(int)
df_train_bal.to_csv("train_balanced.csv", index=False)
print("Saved train_balanced.csv ->", df_train_bal.shape)

# Validation balanced DataFrame
df_val_bal = pd.DataFrame(X_val_bal, columns=feature_names)
df_val_bal['redemption_status'] = y_val_bal.astype(int)
df_val_bal.to_csv("val_balanced.csv", index=False)
print("Saved val_balanced.csv ->", df_val_bal.shape)

# Test processed features (keep id first if present)
df_test = pd.DataFrame(X_test_proc, columns=feature_names)
if test_ids is not None:
    df_test.insert(0, "id", test_ids)
df_test.to_csv("test_features_processed.csv", index=False)
print("Saved test_features_processed.csv ->", df_test.shape)

# Save metadata
meta = {
    'num_original_numeric_columns': len(num_cols),
    'num_original_categorical_columns': len(cat_cols),
    'total_processed_features': len(feature_names),
    'train_balanced_csv': os.path.abspath("train_balanced.csv"),
    'val_balanced_csv': os.path.abspath("val_balanced.csv"),
    'test_features_csv': os.path.abspath("test_features_processed.csv"),
    'preprocessor_path': os.path.abspath("preprocessor.joblib")
}
pd.Series(meta).to_frame("value").to_csv("preprocessing_metadata.csv")
print("Saved preprocessing_metadata.csv")

print("\nAll artifacts created:\n - train_balanced.csv\n - val_balanced.csv\n - test_features_processed.csv\n - preprocessor.joblib\n - preprocessing_metadata.csv")
```

```python
# improved_nn_train_fixed.py
# Run in Colab or local Python environment
# Requirements: torch, sklearn, pandas, numpy, joblib, (optional: imbalanced-learn)
# Install in Colab if needed: !pip install -q imbalanced-learn

import os, time
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, roc_auc_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import joblib

# ---------------- CONFIG ----------------
TRAIN_PATH = "train_balanced.csv"
VAL_PATH   = "val_balanced.csv"
TEST_PATH  = "test_features_processed.csv"
MODEL_OUT  = "coupon_nn_improved_fixed.pth"
SUB_OUT    = "submission_nn_improved_fixed.csv"

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)

BATCH_SIZE = 256
LR = 1e-3
WEIGHT_DECAY = 1e-5
EPOCHS = 80
PATIENCE = 10
PRINT_EVERY = 1

USE_FOCAL = True           # recommended: True
USE_BCE_WITH_LOGITS = False # set True if you want logits + BCEWithLogitsLoss
THRESH_SEARCH = np.linspace(0.05, 0.95, 19)

# ---------------- UTILS ----------------
def load_xy(path):
    df = pd.read_csv(path)
    if 'redemption_status' not in df.columns:
        raise ValueError("redemption_status missing in " + path)
    X = df.drop(columns=['redemption_status']).values.astype(np.float32)
    y = df['redemption_status'].values.astype(np.int64)
    return X, y

def load_test(path):
    df = pd.read_csv(path)
    if 'id' in df.columns:
        ids = df['id'].values
        X = df.drop(columns=['id']).values.astype(np.float32)
    else:
        ids = np.arange(len(df))
        X = df.values.astype(np.float32)
    return X, ids

# ---------------- FOCAL LOSS ----------------
class FocalLoss(nn.Module):
    # for binary classification - inputs are probs (0..1) when logits=False
    def __init__(self, alpha=0.25, gamma=2.0, logits=False, reduction='mean'):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.logits = logits
        self.reduction = reduction

    def forward(self, inputs, targets):
        # inputs: (N,1) floats. targets: (N,1) floats 0/1
        targets = targets.view(-1,1)
        if self.logits:
            # use BCE with logits
            bce_loss = nn.functional.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
            probs = torch.sigmoid(inputs)
        else:
            bce_loss = nn.functional.binary_cross_entropy(inputs, targets, reduction='none')
            probs = inputs
        p_t = probs * targets + (1 - probs) * (1 - targets)
        alpha_factor = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        modulating_factor = (1.0 - p_t) ** self.gamma
        loss = alpha_factor * modulating_factor * bce_loss
        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        return loss

# ---------------- MODEL ----------------
class CouponNN(nn.Module):
    def __init__(self, input_dim, use_logits=False):
        super().__init__()
        self.use_logits = use_logits
        self.net = nn.Sequential(
            nn.Linear(input_dim, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(0.40),

            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.35),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.25),

            nn.Linear(256, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Linear(64, 1)
        )
    def forward(self, x):
        out = self.net(x)
        if self.use_logits:
            return out  # logits
        return torch.sigmoid(out)  # probabilities

# ---------------- LOAD DATA ----------------
print("Loading train/val/test CSVs...")
X_train, y_train = load_xy(TRAIN_PATH)
X_val, y_val = load_xy(VAL_PATH)
X_test, test_ids = load_test(TEST_PATH)

print("Shapes:", X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape)

# Convert to tensors & dataloaders
train_ds = TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train).float().view(-1,1))
val_ds = TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val).float().view(-1,1))
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, drop_last=False)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, drop_last=False)

# ---------------- MODEL + LOSS + OPT ----------------
model = CouponNN(input_dim=X_train.shape[1], use_logits=USE_BCE_WITH_LOGITS).to(DEVICE)

if USE_FOCAL:
    criterion = FocalLoss(alpha=0.25, gamma=2.0, logits=USE_BCE_WITH_LOGITS)
else:
    if USE_BCE_WITH_LOGITS:
        # pos_weight option can be used, but we've balanced data — keep default
        criterion = nn.BCEWithLogitsLoss()
    else:
        criterion = nn.BCELoss()

optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

# ---------------- TRAIN LOOP with early stopping ----------------
best_val_loss = 1e9
patience = 0
best_epoch = -1
start_time = time.time()

for epoch in range(1, EPOCHS+1):
    model.train()
    train_loss = 0.0
    for xb, yb in train_loader:
        xb = xb.to(DEVICE).float()
        yb = yb.to(DEVICE).float()
        optimizer.zero_grad()
        out = model(xb)
        # Ensure shapes and loss function compatibility
        if USE_BCE_WITH_LOGITS:
            loss = criterion(out.view(-1,1), yb)
        else:
            loss = criterion(out, yb)
        loss.backward()
        optimizer.step()
        train_loss += loss.item() * xb.size(0)
    train_loss = train_loss / len(train_loader.dataset)

    # validation
    model.eval()
    val_loss = 0.0
    val_probs = []
    with torch.no_grad():
        for xb, yb in val_loader:
            xb = xb.to(DEVICE).float()
            yb = yb.to(DEVICE).float()
            out = model(xb)
            if USE_BCE_WITH_LOGITS:
                loss = criterion(out.view(-1,1), yb)
                probs = torch.sigmoid(out).cpu().numpy().flatten()
            else:
                loss = criterion(out, yb)
                probs = out.cpu().numpy().flatten()
            val_loss += loss.item() * xb.size(0)
            val_probs.append(probs)
    val_loss = val_loss / len(val_loader.dataset)
    val_probs = np.concatenate(val_probs)

    # scheduler step + print if LR changed
    old_lr = optimizer.param_groups[0]['lr']
    scheduler.step(val_loss)
    new_lr = optimizer.param_groups[0]['lr']
    if new_lr != old_lr:
        print(f" LR reduced from {old_lr:.6f} → {new_lr:.6f}")

    print(f"Epoch {epoch:03d} | Train Loss: {train_loss:.5f} | Val Loss: {val_loss:.5f}")

    # early stopping
    if val_loss < best_val_loss - 1e-6:
        best_val_loss = val_loss
        best_epoch = epoch
        patience = 0
        torch.save(model.state_dict(), MODEL_OUT + ".tmp")
    else:
        patience += 1
        if patience >= PATIENCE:
            print(f"Early stopping at epoch {epoch}. Best epoch was {best_epoch}")
            break

# load best model if saved
if os.path.exists(MODEL_OUT + ".tmp"):
    model.load_state_dict(torch.load(MODEL_OUT + ".tmp", map_location=DEVICE))
    os.replace(MODEL_OUT + ".tmp", MODEL_OUT)
    print("Loaded best model and saved to", MODEL_OUT)

elapsed = time.time() - start_time
print(f"Training finished in {elapsed/60:.2f} minutes. Best val loss: {best_val_loss:.6f} (epoch {best_epoch})")

# ---------------- THRESHOLD TUNING on validation (maximize macro F1) ----------------
model.eval()
with torch.no_grad():
    if USE_BCE_WITH_LOGITS:
        val_probs_all = torch.sigmoid(model(torch.from_numpy(X_val).to(DEVICE).float())).cpu().numpy().flatten()
    else:
        val_probs_all = model(torch.from_numpy(X_val).to(DEVICE).float()).cpu().numpy().flatten()

best_thresh = 0.5
best_macro_f1 = -1.0
for t in THRESH_SEARCH:
    preds = (val_probs_all > t).astype(int)
    m = f1_score(y_val, preds, average='macro')
    if m > best_macro_f1:
        best_macro_f1 = m
        best_thresh = t

print(f"Best threshold (macro-F1): {best_thresh:.3f} with macro F1 = {best_macro_f1:.4f}")

# final metrics at best threshold
val_preds = (val_probs_all > best_thresh).astype(int)
print("\nValidation classification report (tuned threshold):")
print(classification_report(y_val, val_preds, digits=4, zero_division=0))
print("ROC AUC:", roc_auc_score(y_val, val_probs_all))

# ---------------- PREDICT TEST & SAVE ----------------
model.eval()
with torch.no_grad():
    if USE_BCE_WITH_LOGITS:
        test_probs = torch.sigmoid(model(torch.from_numpy(X_test).to(DEVICE).float())).cpu().numpy().flatten()
    else:
        test_probs = model(torch.from_numpy(X_test).to(DEVICE).float()).cpu().numpy().flatten()
test_preds = (test_probs > best_thresh).astype(int)

if test_ids is None:
    submission_df = pd.DataFrame({"id": np.arange(len(test_preds)), "redemption_status": test_preds})
else:
    submission_df = pd.DataFrame({"id": test_ids, "redemption_status": test_preds})
submission_df.to_csv(SUB_OUT, index=False)
print("Saved submission:", SUB_OUT)

print("Done.")

```

```python
# blend_nn_xgb_full_fixed.py
# End-to-end: compute NN val probs -> train XGBoost -> blend -> evaluate -> predict on test
# Requirements: numpy, pandas, torch, xgboost, scikit-learn
# Install if needed: pip install torch xgboost scikit-learn pandas numpy

import os
import time
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report, f1_score, roc_auc_score, accuracy_score
import xgboost as xgb
import torch
import torch.nn as nn

# -------- Config (update file names if needed) --------
TRAIN_CSV = "train_balanced.csv"
VAL_CSV   = "val_balanced.csv"
TEST_CSV  = "test_features_processed.csv"

NN_MODEL_PATH = "coupon_nn_improved_fixed.pth"   # path to your saved PyTorch NN
XGB_MODEL_PATH = "xgb_model.json"

NN_VAL_PROBS_PATH = "nn_val_probs.npy"
XGB_VAL_PROBS_PATH = "xgb_val_probs.npy"

OUTPUT_SUBMISSION = "submission_blend_final.csv"

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)

# -------- Helpers to load CSVs --------
def load_xy(path):
    df = pd.read_csv(path)
    X = df.drop(columns=["redemption_status"]).values.astype(np.float32)
    y = df["redemption_status"].values.astype(np.int64)
    return X, y

def load_test(path):
    df = pd.read_csv(path)
    if "id" in df.columns:
        ids = df["id"].values
        X = df.drop(columns=["id"]).values.astype(np.float32)
    else:
        ids = np.arange(len(df))
        X = df.values.astype(np.float32)
    return X, ids

# -------- Define NN architecture (must match saved NN) --------
class CouponNN(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(0.40),

            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.35),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.25),

            nn.Linear(256, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Linear(64, 1),
            nn.Sigmoid()   # your saved model used sigmoid in forward
        )

    def forward(self, x):
        return self.net(x)

# -------- 1) Load datasets --------
print("Loading datasets...")
X_train, y_train = load_xy(TRAIN_CSV)
X_val, y_val = load_xy(VAL_CSV)
X_test, test_ids = load_test(TEST_CSV)
print("Shapes -> train:", X_train.shape, y_train.shape, "val:", X_val.shape, y_val.shape, "test:", X_test.shape)

# -------- 2) Compute NN validation probabilities (if not saved) --------
def compute_nn_val_probs(nn_model_path, X_val, save_path=NN_VAL_PROBS_PATH, batch_size=4096):
    print("Computing NN validation probabilities...")
    input_dim = X_val.shape[1]
    model = CouponNN(input_dim).to(DEVICE)
    if not os.path.exists(nn_model_path):
        raise FileNotFoundError(f"NN model file not found at {nn_model_path}")
    model.load_state_dict(torch.load(nn_model_path, map_location=DEVICE))
    model.eval()
    probs = []
    with torch.no_grad():
        for i in range(0, X_val.shape[0], batch_size):
            chunk = torch.from_numpy(X_val[i:i+batch_size]).to(DEVICE).float()
            out = model(chunk).cpu().numpy().flatten()
            probs.append(out)
    probs = np.concatenate(probs)
    np.save(save_path, probs)
    print("Saved NN val probs to", save_path)
    return probs

if os.path.exists(NN_VAL_PROBS_PATH):
    nn_val_probs = np.load(NN_VAL_PROBS_PATH)
    if nn_val_probs.shape[0] != X_val.shape[0]:
        print("Existing NN val probs shape mismatch, recomputing.")
        nn_val_probs = compute_nn_val_probs(NN_MODEL_PATH, X_val)
    else:
        print("Loaded cached NN val probs from", NN_VAL_PROBS_PATH)
else:
    nn_val_probs = compute_nn_val_probs(NN_MODEL_PATH, X_val)

# -------- 3) Train XGBoost (fixed function) and get val probabilities --------
def train_xgb_and_get_val(X_train, y_train, X_val, y_val, save_model_path=XGB_MODEL_PATH, save_val_probs=XGB_VAL_PROBS_PATH):
    print("Training XGBoost...")
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'auc',
        'eta': 0.05,
        'max_depth': 7,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'min_child_weight': 3,
        'lambda': 1.0,
        'seed': SEED,
        'verbosity': 1
    }

    num_boost_round = 2000
    early_stopping_rounds = 50
    watchlist = [(dtrain, 'train'), (dval, 'eval')]

    t0 = time.time()
    bst = xgb.train(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        evals=watchlist,
        early_stopping_rounds=early_stopping_rounds,
        verbose_eval=25
    )
    elapsed = (time.time() - t0) / 60.0
    print(f"XGBoost trained in {elapsed:.2f} minutes. Best iteration: {getattr(bst, 'best_iteration', 'N/A')}")

    # Save model
    bst.save_model(save_model_path)
    print("Saved XGB model to", save_model_path)

    # Predict on validation set (bst.predict handles best iteration internally)
    val_probs = bst.predict(dval)
    np.save(save_val_probs, val_probs)
    print("Saved XGB val probs to", save_val_probs)

    return bst, val_probs

if os.path.exists(XGB_VAL_PROBS_PATH):
    xgb_val_probs = np.load(XGB_VAL_PROBS_PATH)
    if xgb_val_probs.shape[0] != X_val.shape[0]:
        print("Existing XGB val probs shape mismatch, retraining XGB.")
        bst, xgb_val_probs = train_xgb_and_get_val(X_train, y_train, X_val, y_val)
    else:
        print("Loaded cached XGB val probs from", XGB_VAL_PROBS_PATH)
        # still need bst for test predictions
        bst = xgb.Booster()
        if os.path.exists(XGB_MODEL_PATH):
            bst.load_model(XGB_MODEL_PATH)
        else:
            # retrain to produce a model file
            bst, xgb_val_probs = train_xgb_and_get_val(X_train, y_train, X_val, y_val)
else:
    bst, xgb_val_probs = train_xgb_and_get_val(X_train, y_train, X_val, y_val)

# -------- 4) Blend search: weight for NN and threshold to maximize macro-F1 --------
print("Searching best blend weight and threshold (maximize macro-F1 on val)...")
best_macro = -1.0
best_config = None

weights = np.linspace(0.0, 1.0, 21)   # 0.0..1.0 step 0.05
thresholds = np.linspace(0.05, 0.95, 19)

for w in weights:
    blended = w * nn_val_probs + (1.0 - w) * xgb_val_probs
    for t in thresholds:
        preds = (blended > t).astype(int)
        m = f1_score(y_val, preds, average='macro')
        if m > best_macro:
            best_macro = m
            best_config = (w, t, m)

w_best, t_best, best_macro = best_config
print(f"Best blend -> nn_weight: {w_best:.3f}, threshold: {t_best:.3f}, macro-F1: {best_macro:.4f}")

# Evaluate blended predictions on validation
blended_best = w_best * nn_val_probs + (1.0 - w_best) * xgb_val_probs
preds_best = (blended_best > t_best).astype(int)
print("\nValidation metrics for blended model (tuned):")
print(classification_report(y_val, preds_best, digits=4, zero_division=0))
print("Accuracy:", accuracy_score(y_val, preds_best))
print("ROC-AUC:", roc_auc_score(y_val, blended_best))

# -------- 5) Create final blended predictions on test and save submission --------
print("Generating blended test predictions...")
# XGBoost test probs
dtest = xgb.DMatrix(X_test)
xgb_test_probs = bst.predict(dtest)

# NN test probs
print("Computing NN test probabilities...")
input_dim = X_test.shape[1]
nn_model = CouponNN(input_dim).to(DEVICE)
nn_model.load_state_dict(torch.load(NN_MODEL_PATH, map_location=DEVICE))
nn_model.eval()
batch = 4096
nn_test_probs = []
with torch.no_grad():
    for i in range(0, X_test.shape[0], batch):
        chunk = torch.from_numpy(X_test[i:i+batch]).to(DEVICE).float()
        out = nn_model(chunk).cpu().numpy().flatten()
        nn_test_probs.append(out)
nn_test_probs = np.concatenate(nn_test_probs)

# blended test probabilities and predictions
test_blend_probs = w_best * nn_test_probs + (1.0 - w_best) * xgb_test_probs
test_blend_preds = (test_blend_probs > t_best).astype(int)

# Save submission
submission = pd.DataFrame({"id": test_ids, "redemption_status": test_blend_preds})
submission.to_csv(OUTPUT_SUBMISSION, index=False)
print("Saved blended submission to", OUTPUT_SUBMISSION)

# Save probs for inspection
np.save("test_nn_probs.npy", nn_test_probs)
np.save("test_xgb_probs.npy", xgb_test_probs)
np.save("test_blend_probs.npy", test_blend_probs)

print("Done. Best blend config:", best_config)
```

```python
# test_predict_helpers.py
# Requirements: torch, pandas, numpy, joblib, sklearn
# Usage: import this module or run as script

import os
import numpy as np
import pandas as pd
import torch
import joblib

# ---------------- CONFIG ----------------
MODEL_PATH = "coupon_nn_improved_fixed.pth"      # path to saved torch model
PREPROCESSOR_PATH = "preprocessor.joblib"       # path to saved ColumnTransformer
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Model class must match architecture used when training:
import torch.nn as nn
class CouponNN(nn.Module):
    def __init__(self, input_dim, use_logits=False):
        super().__init__()
        self.use_logits = use_logits
        self.net = nn.Sequential(
            nn.Linear(input_dim, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(0.40),

            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.35),

            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.25),

            nn.Linear(256, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.15),

            nn.Linear(64, 1)
        )

    def forward(self, x):
        out = self.net(x)
        # If model was saved to output logits set use_logits=True and handle separately.
        # We assume model saved returns logits if you used USE_BCE_WITH_LOGITS = True previously.
        # But most of our scripts used sigmoid on forward (use_logits=False).
        return torch.sigmoid(out)

# ---------------- HELPERS ----------------
def load_preprocessor(path=PREPROCESSOR_PATH):
    if not os.path.exists(path):
        raise FileNotFoundError(f"Preprocessor file not found at {path}")
    return joblib.load(path)

def load_model(model_path=MODEL_PATH, input_dim=None):
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"Model file not found at {model_path}")
    # If model class needs input_dim to construct, require it. We'll infer input_dim from preprocessor when possible.
    if input_dim is None:
        # attempt lazy load: we'll create a dummy small model if input_dim missing later
        raise ValueError("Please provide input_dim (number of processed features) to load the model.")
    model = CouponNN(input_dim=input_dim, use_logits=False)  # change use_logits=True if your saved model outputs logits
    model.load_state_dict(torch.load(model_path, map_location=DEVICE))
    model.to(DEVICE)
    model.eval()
    return model

def preprocess_row(raw_row, preprocessor):
    """
    raw_row: pd.Series, pd.DataFrame (one-row), or dict
    preprocessor: loaded ColumnTransformer
    returns: 1D numpy array (processed features)
    """
    if isinstance(raw_row, dict):
        raw_row = pd.DataFrame([raw_row])
    elif isinstance(raw_row, pd.Series):
        raw_row = raw_row.to_frame().T
    elif isinstance(raw_row, pd.DataFrame):
        if raw_row.shape[0] != 1:
            raise ValueError("preprocess_row expects a single-row DataFrame or Series or dict")
    else:
        raise ValueError("raw_row must be dict, pd.Series, or single-row pd.DataFrame")

    # Align columns: preprocessor was fit on training X_all DataFrame columns
    # If some columns are missing in raw_row, add them with NaN
    expected_cols = preprocessor.feature_names_in_
    for c in expected_cols:
        if c not in raw_row.columns:
            raw_row[c] = np.nan
    # ensure column order
    raw_row = raw_row[expected_cols]
    X_proc = preprocessor.transform(raw_row)  # returns ndarray
    # If result is sparse matrix or 2D array, convert to 1d numpy array
    if hasattr(X_proc, "toarray"):
        X_proc = X_proc.toarray()
    return X_proc.flatten()

def predict_single_processed(model, processed_vector, threshold=0.5):
    """
    processed_vector: 1D numpy array of processed features (length == input_dim)
    model: loaded torch model
    returns: prob (float), pred_class (int)
    """
    model.eval()
    x = torch.tensor(processed_vector, dtype=torch.float32).unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        prob = model(x).cpu().numpy().flatten()[0]  # model returns probability (sigmoid applied in forward)
    pred = int(prob > threshold)
    return float(prob), pred

def predict_single_raw(model, preprocessor, raw_row, threshold=0.5):
    """
    raw_row: pd.Series / single-row DataFrame / dict with raw columns like in X_all used for preprocessing
    preprocessor: loaded ColumnTransformer
    """
    proc = preprocess_row(raw_row, preprocessor)
    return predict_single_processed(model, proc, threshold)

def predict_batch(model, preprocessor, df, threshold=0.5, raw=True):
    """
    df: if raw==True -> DataFrame with raw columns (multiple rows)
        if raw==False -> DataFrame or ndarray with processed features (no 'id' col)
    returns: DataFrame with columns ['id' (if present), 'prob', 'pred']
    """
    if raw:
        # Ensure df contains all expected columns
        expected_cols = preprocessor.feature_names_in_
        missing = [c for c in expected_cols if c not in df.columns]
        if missing:
            # fill missing with NaN
            for c in missing:
                df[c] = np.nan
        df_proc = preprocessor.transform(df[expected_cols])
        if hasattr(df_proc, "toarray"):
            df_proc = df_proc.toarray()
    else:
        # df is processed features
        if isinstance(df, pd.DataFrame):
            df_proc = df.values
        else:
            df_proc = np.asarray(df)
    # batch predict
    model.eval()
    probs = []
    batch = 1024
    with torch.no_grad():
        for i in range(0, df_proc.shape[0], batch):
            chunk = torch.tensor(df_proc[i:i+batch], dtype=torch.float32).to(DEVICE)
            out = model(chunk).cpu().numpy().flatten()
            probs.append(out)
    probs = np.concatenate(probs)
    preds = (probs > threshold).astype(int)
    # return DataFrame (preserve id if present)
    result = pd.DataFrame({"prob": probs, "pred": preds})
    if 'id' in df.columns:
        result.insert(0, "id", df['id'].values)
    return result

# ---------------- EXAMPLE USAGE ----------------
if __name__ == "__main__":
    print("Loading preprocessor...")
    preproc = load_preprocessor(PREPROCESSOR_PATH)
    input_dim = len(preproc.get_feature_names_out())
    print("Processed feature count (input_dim):", input_dim)

    print("Loading model...")
    model = load_model(MODEL_PATH, input_dim=input_dim)

    # example 1: predict first row from processed test file
    test_df = pd.read_csv("test_features_processed.csv")
    if 'id' in test_df.columns:
        print("Example: using processed test file row 0")
        proc_row = test_df.drop(columns=['id']).iloc[0].values.astype(np.float32)
        prob, pred = predict_single_processed(model, proc_row, threshold=0.5)
        print("Processed sample -> prob:", prob, "pred:", pred)
    else:
        print("test_features_processed.csv has no id column — assuming all columns are feature columns")
        proc_row = test_df.iloc[0].values.astype(np.float32)
        prob, pred = predict_single_processed(model, proc_row, threshold=0.5)
        print("Processed sample -> prob:", prob, "pred:", pred)

    # example 2: predict using raw feature values (single row dict)
    # raw_row must contain the original raw feature column names exactly as when preprocessor was fit (preprocessor.feature_names_in_)
    example_raw = {}
    # fill example_raw with some default or real values, e.g.:
    # example_raw = {'campaign_type': 'Y', 'campaign_duration': 30, 'num_unique_items': 10, ...}
    print("You can call predict_single_raw(model, preproc, raw_row_dict, threshold) with a raw-row dict/Series")

    # example 3: batch predict (processed)
    print("Batch predicting on the processed test file and saving predictions to submission_test_preds.csv ...")
    # if the test file is processed with 'id' first, drop id before passing raw=True
    if 'id' in test_df.columns:
        batch_result = predict_batch(model, preproc, test_df, threshold=0.5, raw=True)
    else:
        batch_result = predict_batch(model, preproc, test_df, threshold=0.5, raw=False)
    batch_result.to_csv("test_predictions_batch.csv", index=False)
    print("Wrote test_predictions_batch.csv with shape", batch_result.shape)
```
